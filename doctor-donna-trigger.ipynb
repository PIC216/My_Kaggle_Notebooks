# Doctor Donna Trigger

## "Westerly. Pelican. Dreams. Tornado. Clifftops. Andante. Grief. Fingerprint. Susurration. Sparrow. Dance. Mexico. Binary..."

In this notebook I want to investigate these words and explore the relevance of these trigger words which bring the Doctor Donna back to our screens.

For those of you who don't love Doctor Who, Donna was the Doctor's companion during the 2008 series. She was forced to forget the Doctor when she experienced a human / time lord metacrises which gave her (a human) the knowledge of the doctor (a time lord), but humans cannot hold the knowledge of a time lord therefore her memories had to be wiped. Donna came back in the 60th anniversary special and her time lord knowledge was required so the doctor started listing seemingly random trigger words which eventually bring back the beloved Doctor Donna

I'll start by assigning the words to a list.

```python
# create list
trigger = ["Westerly", "Pelican", "Dreams", "Tornado", "Clifftops", "Andante", "Grief", "Fingerprint",
           "Susurration", "Sparrow", "Dance", "Mexico", "Binary"]
```

### Episode titles
Now let's pull in details of Doctor Who episodes to see if any of these words appear in any episode names.

```python
def check_for_word(word, df, col):
    '''
    This function checks for a specific word within a column of a dataframe.
    It takes 3 arguments: 
      the word,
      the dataframe,
      the relevant column of the dataframe.
    It returns a list of 3 items:
      whether the word exists in the dataframe column,
      the number of times the word appears,
      the dataframe filtered to only include rows where the word appears
    '''
    filtered_df = df[df[col].str.contains(word, case = False, na=False)] # filter the dataframe for relevant words
    num_rows = filtered_df.shape[0] # calculate the number of rows in the dataframe
    if num_rows == 0: # state whether the word appears in the dataframe
        word_exists = False
    else:
        word_exists = True
    # return tuple containing
    #  a boolean for whether the word exists,
    #  an int for the number of times the word appears,
    #  a dataframe containing rows where the word is included
    return word_exists, num_rows, filtered_df
```

```python
# Pull in DW details dataset
dw_episode_details = pd.read_csv('../input/doctor-who/all-detailsepisodes.csv')

for word in trigger: # loop through each word in the trigger
    word_exists = check_for_word(word, dw_episode_details, "title")[0] # run function and assign only whether the word appears in the dataframe
    if word_exists: # print statement regarding whether or not the word appears
        print(f"The word {word} is in an episode title!")
    else:
        print(f"The word {word} is not in an episode title...")
```

Yay, one of the words is in the title of at least one episode! Though, I had expected more to appear in episode titles...

Let's see which episode titles the word Dance appears in.

```python
dance_episode = check_for_word("Dance", dw_episode_details, "title")[2] # run function and assign only the filtered dataframe
dance_episode # view dataframe
```

Well the word "Dance" appears in two episodes. One with the 5th doctor and one with the 9th doctor.

As the 11th word I was hoping it would appear in an episode with the 11th doctor, but since it didn't I'm going to assume the word appearing in an episode title was more likely a coincidence, so I will continue my search for the significance of the words.

### Episode scripts

Now let's pull in all of the Doctor Who scripts to see how often these words appear in any lines.

```python
# Pull in DW script dataset
dw_scripts = pd.read_csv('../input/doctor-who/all-scripts.csv')

for word in trigger: # loop through each word in the trigger
    word_exists, num_rows = check_for_word(word, dw_scripts, "text")[0:2] # run function and assign whether the word appears in the dataframe and how many times
    if word_exists: # print statement regarding whether or not the word appears
        print(f"The word {word} appears {num_rows} times across all doctor who scripts.\n")
    else:
        print(f"The word {word} does not appear in any episode...\n")
```

So, two of the words, Pelican and Andante, do not appear in any previous scripts, while all the others appear from 2 to 200(+) times.
