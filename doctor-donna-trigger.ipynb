# Doctor Donna Trigger

## "Westerly. Pelican. Dreams. Tornado. Clifftops. Andante. Grief. Fingerprint. Susurration. Sparrow. Dance. Mexico..."

In this notebook I want to investigate these words and explore the relevance of these trigger words which bring the Doctor Donna back to our screens.

Note: As you may have noticed, I have ignored "Binary", hopefully if you are looking at this you already know the relevance of the word "binary" as a trigger.

I'll start by assigning the words to a list.

```python
trigger = ["Westerly", "Pelican", "Dreams", "Tornado", "Clifftops", "Andante", "Grief", "Fingerprint", "Susurration", "Sparrow", "Dance", "Mexico"]
```

First, let's check the number of words in the sequence or whether the sequence is some form of mnemonic.

```python
print(len(trigger)) # print the length of the list containing the trigger words

for word in trigger:
    print(word[0]) # print the first letter of each word in the list of trigger words

# I also tried printing the second, third, fourth, and fifth letters; to no avail
# (sixth letter onwards was not attempted as shortest words in the trigger list are 5 letters long)
#
# print("\n")
# for word in trigger:
#     print(word[1])
```
